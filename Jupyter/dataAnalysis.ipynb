<h2>Lib Imports</h2>

```python
import os
import pandas as pd
```

<h2>Variables</h2>

```python
data = pd.read_csv("../CSV/Car_Top_Speed_Table.csv", sep=",")
countriesAvg = dict()
makeAvg = dict()
display(data.head())
```

<h2>Codes</h2>

```python
for make in data['Vehicle Make'].unique():
    makeAvg[make] = None

for country in data['Country '].unique():
    countriesAvg[country] = data.loc[data['Country '] == country]["Top Speed "].mean()

display(countriesAvg)
display(makeAvg)
```
